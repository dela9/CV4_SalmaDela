```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r'/content/Dela.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img2 = img.copy()

img2 = cv2.rectangle(img2, (420, 760), (520, 860), color=(0, 0, 255), thickness=3)
img2 = cv2.rectangle(img2, (540, 760), (620, 860), color=(0, 0, 255), thickness=3)
img2 = cv2.rectangle(img2, (350, 630), (650, 980), color=(255, 0, 0), thickness=5)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img2, text="face", org=(350, 600), fontFace=font, fontScale=2, color=(255, 0, 0), thickness=3)

cv2.putText(img2, text="eye", org=(400, 730), fontFace=font, fontScale=2, color=(0, 0, 255), thickness=3)
cv2.putText(img2, text="eye", org=(530, 730), fontFace=font, fontScale=2, color=(0, 0, 255), thickness=3)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img)
axes[0].set_title('Normal')

axes[1].imshow(img2)
axes[1].set_title('Marking')

plt.show()
```
